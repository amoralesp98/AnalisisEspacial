```python
# Columnas necesarias para el trabajo
minedu=minedu[['ID', 'CEN_EDU', 'D_NIV_MOD', 'TALUM_HOM','TALUM_MUJ','TALUMNO','X_COORD','Y_COORD','DEPARTAMEN', 'PROVINCIA','DISTRITO' ]]
```

```python
minedu.head(3)
```

```python
#Filtrar la capa minedu por dos campos distintos
minedu_miraflores = minedu[(minedu['DEPARTAMEN'] == 'LIMA') &  (minedu['PROVINCIA'] == 'LIMA')& (minedu['DISTRITO'] == 'MIRAFLORES')]
```

```python
minedu_miraflores.sample(2)
```

Generacion de Graficas usando Librerias de Plotly

```python
fig=px.scatter_mapbox(
    minedu_miraflores,
    lat='Y_COORD',
    lon='X_COORD',
    mapbox_style='carto-positron',
    zoom=13,
    hover_name='CEN_EDU',
    color='TALUMNO',
    color_continuous_scale='matter',
    width=800,

)
fig.update_traces(
    marker=dict(size=10)
)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
#Crear un papa de burbujas
fig=px.scatter_mapbox(
    minedu_miraflores,
    lat='Y_COORD',
    lon='X_COORD',
    mapbox_style='carto-positron',
    zoom=13,
    hover_name='D_NIV_MOD',
    color='D_NIV_MOD',
    size='TALUMNO',
    size_max=20,
    width=1000
)

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

## Visualizacion: HeatMap

```python
fig=px.density_mapbox(
    minedu_miraflores,
    lat='Y_COORD',
    lon='X_COORD',
    mapbox_style='carto-positron',
    color_continuous_scale='inferno',
    zoom=13,
    radius=15,
    opacity=0.8,
    width=800,
    #height=500,
)
fig.update_layout(
    margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
